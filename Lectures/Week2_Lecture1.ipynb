# Introduction

## Data: collection of numbers, characters, images, videos, or any other items that provide information about something

## Statistics: Helps to make sense of the Data ( often by using sampling from a population)



# Central Values ( Numerical )
## Median: The center of the data
## Mean : Arithmatic average
## Mode: Most frequent, Nonunique,  sometimes does not exist
## Locating the center is only part of the story!
## These values do not give you the full detail of the data and are misleading sometime

# Spreads ( Variability, Disperson)
## Range  =  Maximum  –  Minimum
## Percentile:  Divides the data in one hundred groups.  
#### First quartile (Q1) is the 25th percentile, Median is 50th percentile, Third quartile(Q3) is the 75th percentile
#### 1 quartile = 0.25 quantile = 25 percentile
## Interquartile Range (IQR) = Q3 – Q1

# More on Spreads: Variance and Standard Deviation
# Data: (x1, x2, xn)
## Variance (Population): $\displaystyle{\large\sigma^2 = \dfrac{\sum_{i=1}^n(x_i - \bar{x})^2}{n}}$
## Variance (Sample) : $\displaystyle{\large s^2 = \dfrac{\sum_{i=1}^n(x_i - \bar{x})^2}{n-1}}$
## Standard Deviation (Population) : $\displaystyle{\large\sigma = \sqrt{\dfrac{\sum_{i=1}^n(x_i - \bar{x})^2}{n}}}$
## Standard Deviation (Sample) :  $\displaystyle{\large s =\sqrt{ \dfrac{\sum_{i=1}^n(x_i - \bar{x})^2}{n-1}}}$
## Bigger $\large s$ or $\large \sigma \implies $ Data is more spread from the mean $\large \bar{x}$

```python
# Anscome's Quartlet
y1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
y2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
y4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]

x = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
x4 = [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8]
```

```python
import numpy as np
print("Means:", np.mean(y1),np.mean(y2), np.mean(y2), np.mean(y3), np.mean(y4))
```

```python
# Anscombe's quartlet
#https://matplotlib.org/stable/gallery/specialty_plots/anscombe.html
import matplotlib.pyplot as plt

datasets = {
    'I': (x, y1),
    'II': (x, y2),
    'III': (x, y3),
    'IV': (x4, y4)
}

fig, axs = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(16, 16), 
                        gridspec_kw={'wspace': 0.08, 'hspace': 0.08})
axs[0, 0].set(xlim=(0, 20), ylim=(2, 14))
axs[0, 0].set(xticks=(0, 10, 20), yticks=(4, 8, 12))

for ax, (label, (x, y)) in zip(axs.flat, datasets.items()):
    ax.text(0.1, 0.9, label, fontsize=20, transform=ax.transAxes, va='top')
    ax.tick_params(direction='in', top=True, right=True)
    ax.plot(x, y, 'o')

    # linear regression
    p1, p0 = np.polyfit(x, y, deg=1)  # slope, intercept
    ax.axline(xy1=(0, p0), slope=p1, color='r', lw=2)

    # add text box for the statistics
    stats = (f'$\\mu$ = {np.mean(y):.2f}\n'
             f'$\\sigma$ = {np.std(y):.2f}\n'
             f'$r$ = {np.corrcoef(x, y)[0][1]:.2f}')
    bbox = dict(boxstyle='round', fc='blanchedalmond', ec='orange', alpha=0.5)
    ax.text(0.95, 0.07, stats, fontsize=9, bbox=bbox,
            transform=ax.transAxes, horizontalalignment='right')
plt.show()
```

# Rolling a Six-Sided Die 6,000,000 Times
### If randrange truly produces integers at random, every number in its range has an equal probability (or chance or likelihood) of being returned each time we call it.
### Roll a die 6,000,000 times.
### Each die face should occur approximately 1,000,000 times.
### We used Python’s underscore (_) digit separator to make the value 6000000 more readable.

```python
frequency1 = 0 # face frequency counters
frequency2 = 0
frequency3 = 0
frequency4 = 0
frequency5 = 0
frequency6 = 0

for roll in range(6_000_000):  # 6,000,000 die rolls. Note underscore separators
    face = random.randrange(1, 7)                          
    if face == 1:
        frequency1 += 1 # increment appropriate face counter
    elif face == 2:
        frequency2 += 1
    elif face == 3:
        frequency3 += 1
    elif face == 4:
        frequency4 += 1
    elif face == 5:
        frequency5 += 1
    elif face == 6:
        frequency6 += 1
```

```python
# Read more about f-string. https://peps.python.org/pep-0498/#how-to-denote-f-strings
print(f'Face{"Frequency":>13}')
print(f'{1:>4}{frequency1:>13}')
print(f'{2:>4}{frequency2:>13}')
print(f'{3:>4}{frequency3:>13}')
print(f'{4:>4}{frequency4:>13}')
print(f'{5:>4}{frequency5:>13}')
print(f'{6:>4}{frequency6:>13}')
```

## Some Probability Distributions

## Uniform Distribution
###  Uniform distribution is a probability distribution where all outcomes are equally likely.

```python
from scipy.stats import *
a, b = 0.2, 0.8
x = np.linspace(0, 5, 1000)   
y1 = uniform.pdf(x, a, b) 
y2 = uniform.pdf(x, a, b) 
plt.plot(x, y1, "o", x, y2, "r--") 
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the rate parameter
lambda_param = 0.5

# Generate random samples from an exponential distribution
samples = np.random.exponential(scale=1/lambda_param, size=1000)

# Plot the histogram of the samples
plt.hist(samples, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')

# Plot the theoretical probability density function (PDF)
x = np.linspace(0, 10, 100)
pdf = lambda_param * np.exp(-lambda_param * x)
plt.plot(x, pdf, label='Theoretical PDF', color='red')

plt.title('Exponential Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

## Beta and Gamma Distribution

* Beta distributions are often employed in business settings for modeling uncertainty and risk, while gamma distributions find utility in areas like project management, reliability analysis, and insurance to capture the distribution of time until events occur.

```python
samples = np.random.beta(2, 5, size=1000)
x = np.linspace(samples.min(), samples.max(), 1000)
y = beta.pdf(x, 2, 5)
plt.hist(samples, alpha=0.2, bins=20, density=True, color= 'b')
plt.plot(x, y)
plt.show()
```

```python
samples = np.random.gamma(1, size=1000)
x = np.linspace(samples.min(), samples.max(), 1000)
y = gamma.pdf(x, 1)
plt.hist(samples, alpha=0.2, bins=20, density=True, color = 'r')
plt.plot(x, y)
plt.show()
```

## Basic Line Plot

```python
from datetime import datetime
x = pd.period_range(datetime.now(), periods=300, freq='d')
x = x.to_timestamp().to_pydatetime() 
y = np.random.randn(300, 3).cumsum(0)  # 3 columns of y values
plt.plot(x, y)
```

## Detailed plot with axes, title, legends
### Saving the plot for later use

```python
plots = plt.plot(x, y[:,0], '-', x, y[:,1], '--', x, y[:,2], '-.')
plt.legend(plots, ('y1', 'y2', 'y3'), loc='best',
                    framealpha=0.25,
                    prop={'size':'small', 'family':'monospace'})
plt.gcf().set_size_inches(8, 4)
plt.title('Random trends')
plt.xlabel('Date')
plt.ylabel('Cumulative sum')
plt.grid(True) 
plt.tight_layout() 
plt.savefig('3randomlines.png')
```

## Subplots

```python
fig, axes = plt.subplots(
                    nrows=3, ncols=1, 
                    sharex=True, sharey=True, 
                    figsize=(8, 8))
labelled_data = zip(y.transpose(), 
                    ('y1', 'y2', 'y3'), ('b', 'g', 'r'))
fig.suptitle('Three Random Trends', fontsize=16)
for i, ld in enumerate(labelled_data):
    ax = axes[i]
    ax.plot(x, ld[0], label=ld[1], color=ld[2])
    ax.set_ylabel('Cumulative sum')
    ax.legend(loc='upper left', framealpha=0.5,
              prop={'size':'small'})
axes[-1].set_xlabel('Date') 
```
