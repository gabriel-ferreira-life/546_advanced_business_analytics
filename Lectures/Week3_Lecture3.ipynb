```python
# Import the package 
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score #
km = KMeans(3)
km_clusters = km.fit_predict(x)
lifex["clusters"] = km_clusters
lifex.head(3)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
standard = StandardScaler()
x_standard = standard.fit_transform(x)
x_standard.shape
```

```python
x_standard
```

```python
# Not scaled
c_nums = list(range(2, 10))
clusters_SSE = []
for k in c_nums:
    kmeans = KMeans(n_clusters=k)
    model = kmeans.fit(x)
    clusters_SSE.append(model.inertia_)
plt.plot(clusters_SSE, marker='o', markersize=10)
```

```python
# Standard Scaler
c_nums = list(range(2, 10))
clusters_SSE = []
for k in c_nums:
    kmeans = KMeans(n_clusters=k)
    model = kmeans.fit(x_standard)
    clusters_SSE.append(model.inertia_)
plt.plot(clusters_SSE, marker='o', markersize=10)
```

```python
km = KMeans(2)
newkm_clusters = km.fit_predict(x_standard)
lifex['newclusters'] = list(newkm_clusters)
lifex.head(5)
```

```python
plt.scatter(x['Life expectancy '], x['Schooling'], c=newkm_clusters)
plt.title("World Country Clusters from K-Means Clustering")
plt.xlabel("Life Expectancy")
plt.ylabel("Schooling")
plt.show()
```

```python
silhouette_score(x, lifex.newclusters)
```


# Anomaly/Outlier/Novelty Detection

# Association Rule MIning

### Point anomalies: A single data point that is significantly different from the rest of the data. This can be caused by measurement errors, data entry errors, or a rare event.

### Contextual anomalies: A data point that is not anomalous by itself, but becomes anomalous when considered in its context. For example, a transaction that is normal for a customer may be anomalous if it occurs at an unusual time or location.

### Collective anomalies: A group of data points that are anomalous when considered together, but not when considered individually. For example, a sudden increase in the number of login attempts may indicate a cyber attack.

### Continuous anomalies: A data point or set of data points that deviate from the expected trend over time. For example, a sudden drop in the stock price of a company may indicate a problem with the company.

## Isolation Forest
#### The Isolation Forest is a technique for the detection of outlier samples that uses a tree.
#### Since outliers have features X that differ significantly from most of the samples, they are isolated earlier in the hierarchy of a decision tree.
#### Outliers are detected by setting a threshold on the mean length (number of splits) from the top of the tree downwards.
#### The Scikit-learn implementation provides a score for each sample that increases from -1 to +1 with the number of splits.

```python
cd.columns
```

```python
cols = ['A1', 'A2', 'A3', 'A8', 'A9', 'A10', 'A11', 'A12', 'A14', 'A15', 'R1']
```

```python
# Import the package 
from sklearn.ensemble import IsolationForest
isf = IsolationForest(n_estimators=25, random_state=0, 
                      contamination= 0.01).fit(cd[cols])
isf_classes = isf.predict(cd[cols])
cd['anomaly']= isf_classes
anomaly = cd[cd.anomaly == -1]
anomaly
```

## Local Outlier Factor ( LOF)
### Local Density: the inverse of the average distance to its k-nearest neighbors.
### Local Outlier Factor: the ratio of the local density of p and those of pâ€™s k-nearest neighbors.
### The LOF of each object depends on the density of the cluster relative to it and the distance between it and the cluster.
https://en.wikipedia.org/wiki/Local_outlier_factor

```python
from sklearn.neighbors import LocalOutlierFactor
lof = LocalOutlierFactor(n_neighbors=25)
ypred = lof.fit_predict(cd[cols])
cd['lof_anomaly']= ypred
anomaly = cd[cd.lof_anomaly == -1]
anomaly
```

## Benford's Law for Fraud Detection
### Benfordâ€™s Law states that the first digits found in a data set are expected to be arranged in a way that the lowest digit, one, appears the most frequently, followed by two, three, etc. This law can be utilized to detect patterns, or lack thereof, in naturally occurring data sets, which can be used to help catch anomalies or fraud in data.

### A set of numbers is said to satisfy Benford's law if the leading digit d (d âˆˆ {1, ..., 9}) occurs with probability

$\large \displaystyle P(d)=\log _{10}(d+1)-\log _{10}(d)=\log _{10}\left({\frac {d+1}{d}}\right)$

```python
import matplotlib.pyplot as plt
from math import log10
x = [i for i in range(1, 10)]
results = [round(log10(n+1) - log10(n),3) for n in x]
fig, ax = plt.subplots()
plt.bar(x, results)
plt.xticks(x, x)
for i, v in enumerate(results):
    plt.text(x[i] - 0.25, v + 0.01, str(v))
plt.show()
```

https://en.wikipedia.org/wiki/Benford%27s_law

```python
import mlxtend
sales = pd.read_csv('salesdata.csv')
sales.head()
```

```python
items = sales[["CustomerID", "Itemname"]]
items = items.astype(str).replace(r'\.0$', '', regex=True)
items['Itemname'] = items['Itemname'].str.lower()
items.head()
```

```python
items.dtypes
```

```python
items.shape
```

```python
itemlist = items.groupby("CustomerID")["Itemname"].agg(list).to_frame().reset_index()
itemlist.head()
```

```python
itemlist.Itemname[1]
```

```python
data = list(itemlist.Itemname)
data[:2]
```

```python
from mlxtend.preprocessing import TransactionEncoder
encoding = TransactionEncoder()
encoded = encoding.fit(data).transform(data)
df = pd.DataFrame( encoded,columns= encoding.columns_)
df.head()
```

```python
from mlxtend.frequent_patterns import apriori
df = apriori(df, min_support = 0.05, use_colnames = True, verbose = 1)
df.head()
```

```python
from mlxtend.frequent_patterns import association_rules
df = association_rules(df, metric = "confidence", min_threshold = 0.4)
df.head()
```
